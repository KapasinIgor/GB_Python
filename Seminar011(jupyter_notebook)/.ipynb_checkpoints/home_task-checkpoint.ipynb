# Дана функция f(x) = 0.6x^3+5.5x^2+10x-5

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *

a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol("x")

func_x = a * (x**3) + b * (x**2) + c * (x**1) + d

deriv = diff(func_x)
```

```python
func_x
```

```python
deriv
```

#### 1. Определяем корни

```python
print("Корень №1:", round(solve(func_x, x)[0].n(), 2))
print("Корень №2:", round(solve(func_x, x)[1].n(), 2))
```

#### 2. Находим интервалы, на которых функция возрастает

```python
solve(deriv > 0, x)
```

#### 3. Находим интервалы, на которых функция убывает

```python
solve(deriv < 0, x)
```

#### 4. Строим графики

```python
from sympy.plotting import plot

g1 = plot(func_x, (x,-4,4), line_color='red', title='график функции')
```

```python
g2 = plot(deriv, (x,-4,4), line_color='orange', title='график производной')
```

```python
g3 = plot(func_x, deriv, (x,-4,4), title = 'Совмещенный график', legend = True, xlabel = 'X', ylabel = 'Y', show = False)
g3[0].line_color = 'red'
g3[1].line_color = 'blue'
g3.show()
```

#### 5. Вычисляем вершину

```python
print("Вершина:", '(', round(-b/(2 * a), 2), ',', func_x.subs(x, (-b/(2 * a))).n(3), ')')
```

#### 6. Определяем промежутки, на котором f > 0

```python
solve(func_x > 0, x)
```

#### 7. Определяем промежутки, на котором f < 0

```python
solve(func_x < 0, x)
```
