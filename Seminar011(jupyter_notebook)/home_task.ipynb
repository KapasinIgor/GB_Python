# Дана функция f(x) = 0.6x^3+5.5x^2+10x-5

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *

a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol("x")

func_x = a * (x**3) + b * (x**2) + c * (x**1) + d

deriv = diff(func_x)
```

```python
func_x
```

```python
deriv
```

#### 1. Определяем корни

```python
print("Корень №1:", solve(func_x, x, cubics=False)[0].n(3))
print("Корень №2:", solve(func_x, x, cubics=False)[1].n(3))
print("Корень №3:", solve(func_x, x, cubics=False)[2].n(3))
```

#### 2. Находим интервалы, на которых функция возрастает

```python
solve(deriv > 0, x)
```

#### 3. Находим интервалы, на которых функция убывает

```python
solve(deriv < 0, x)
```

#### 4. Строим графики

```python
from sympy.plotting import plot

g1 = plot(func_x, (x,-30,30), line_color='red', title='график функции')
```

```python
g2 = plot(deriv, (x,-10,5), line_color='blue', title='график производной')
```

```python
g3 = plot(func_x, deriv, (x,-6.5,2.5), title = 'Совмещенный график(точки пересечения)', legend = True, xlabel = 'X', ylabel = 'Y', show = False)
g3[0].line_color = 'red'
g3[1].line_color = 'blue'
g3.show()
```

#### 5. Вычисляем вершины

```python
max = solve(deriv, x, cubics=False)[0].n(3)
min = solve(deriv, x, cubics=False)[1].n(3)
print("Точки экстремума:", max, min)
```

#### 6. Определяем промежутки, на котором f > 0

```python
solve(func_x > 0, x, cubics=False)
```

#### 7. Определяем промежутки, на котором f < 0

```python
solve(func_x < 0, x, cubics=False)
```
